```python
import yfinance as yf
from datetime import datetime
```

```python
data_hoje = datetime.today().strftime('%Y-%m-%d')
data_hoje
```

```python
# Variaveis
ticker = 'PETR4.SA'#'BTC-USD'
start_data = '2020-01-01'
end_data = data_hoje

# Coletar Dados
df = yf.download(ticker, start = start_data, end = end_data, progress = False)
df = df.xs(ticker, level='Ticker', axis=1)
df
```
